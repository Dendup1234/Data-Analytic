```python
# Handling the missing data value
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
df['oldpeak'] = imputer.fit_transform(df[['oldpeak']])
df['thalch'] = imputer.fit_transform(df[['thalch']])
df['trestbps'] = imputer.fit_transform(df[['trestbps']])
df['chol'] = imputer.fit_transform(df[['chol']])
```

```python
# Handling the missing value of the categorical values
most_frequent_fbs = df['fbs'].value_counts().index[0]
df['fbs'] = df['fbs'].fillna(most_frequent_fbs)

#for the exang
most_frequent_exang = df['exang'].value_counts().index[0]
df['exang'] = df['exang'].fillna(most_frequent_exang)

# for the thal
most_frequent_thal = df['thal'].value_counts().index[0]
df['thal'] = df['thal'].fillna(most_frequent_thal)


# for the ca
most_frequent_ca = df['ca'].value_counts().index[0]
df['ca'] = df['ca'].fillna(most_frequent_ca)

# for the slope
most_frequent_slope = df['slope'].value_counts().index[0]
df['slope'] = df['slope'].fillna(most_frequent_slope)


# for the restecg finding the most occuring categories
most_frequent_restecg = df['restecg'].value_counts().index[0]
df['restecg'] = df['restecg'].fillna(most_frequent_restecg)
```

```python
# conversion of one hot encoding
df = pd.get_dummies(df,columns=['sex'],dtype=int)
```
