```python
import pandas as pd
df1 = pd.read_json("pokedex.json")
df2 = pd.read_csv("Complete_pokedex.csv")
```

```python
null_counts = df2.isnull().sum()
null_counts = null_counts[null_counts > 0]  # Keep only columns with NaNs
print(null_counts)
```

```python
#Handling the null value
df2['evolves_from'] = df2['evolves_from'].fillna("None")
df2['egg_group_2'] = df2['egg_group_2'].fillna("None")
```

```python
df1.columns
```

```python
df2.columns
```

```python
duplicated_columns =set(df1.columns) & set(df2.columns)
duplicated_columns
```

```python
# Replacing the columns name for the consistency
df2.rename(columns={"pokemon_name":"Name"},inplace = True)
```

```python
df1.drop(columns=['Id',],inplace= True)
```

```python
#Removing the duplicated attributes for the consistency of the data sets
df2.drop(columns=['type_1','type_2','ability_1','ability_2','ability_3','height','weight','bmi','hit_points','attack','defense','special_attack','special_defense',
                 'speed'],inplace= True)
```

```python
#Now merging the both the data frame
merged_df = pd.merge(df1,df2,on="Name",how="inner")
```

```python
merged_df
```

```python
merged_df.columns
```

```python
print(merged_df.head())
```

```python
# To bring the Id column in the intial column in the given dataframe
merged_df = merged_df.reindex(columns=['pokedex_number'] + [col for col in merged_df.columns if col != 'pokedex_number'])
```

```python
merged_df.to_csv("processed_data.csv", index=False)
```

```python
# Heatmap of Type Effectiveness
import matplotlib.pyplot as plt
import seaborn as sns
type_columns = [col for col in merged_df.columns if "against_" in col]
type_effectiveness = merged_df[type_columns].mean().reset_index()
type_effectiveness.columns = ["Type", "Effectiveness"]
plt.figure(figsize=(12, 5))
sns.heatmap(merged_df[type_columns].corr(), cmap="coolwarm", annot=True)
plt.title("Pokémon Type Effectiveness Heatmap")
plt.show()
```

```python
# 3. Bar Plot of Primary Pokémon Types
plt.figure(figsize=(10, 5))
sns.countplot(y=merged_df["Type 1"], order=merged_df["Type 1"].value_counts().index)
plt.title("Frequency of Primary Pokémon Types")
plt.show()
```
